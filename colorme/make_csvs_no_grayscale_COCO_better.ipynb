```python
import os
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from tqdm import tqdm
import torchvision.transforms as transforms
```

```python
train_directory = '/datasets/COCO-2015/train2014'
val_directory = '/datasets/COCO-2015/val2014'
test_directory = '/datasets/COCO-2015/test2015'
```

# Training set

```python
image_list = []
for image in tqdm(os.listdir(train_directory)):
    img_name = os.path.join(train_directory, image)
    img = Image.open(img_name)
    np_im = np.array(img)
    dims = np_im.shape
    if len(dims) == 3 and dims[0] >= 128 and dims[1] >= 128:
        image_list.append(img_name)
    elif dims[0] < 64 or dims[1] < 64:
        print(image, dims)
print(len(image_list))
```

```python
train_s = pd.Series(image_list, name='images')
train_s.to_csv('train_COCO_no_grayscale_better.csv', header = True, index = False)
```

```python
# Test a few
tests = np.random.randint(0, len(image_list), 10)
fig, ax = plt.subplots(ncols = 10, figsize = (20, 10))
ax[0].imshow(Image.open(train_s[tests[0]]))
ax[1].imshow(Image.open(train_s[tests[1]]))
ax[2].imshow(Image.open(train_s[tests[2]]))
ax[3].imshow(Image.open(train_s[tests[3]]))
ax[4].imshow(Image.open(train_s[tests[4]]))
ax[5].imshow(Image.open(train_s[tests[5]]))
ax[6].imshow(Image.open(train_s[tests[6]]))
ax[7].imshow(Image.open(train_s[tests[7]]))
ax[8].imshow(Image.open(train_s[tests[8]]))
ax[9].imshow(Image.open(train_s[tests[9]]))
plt.tight_layout()
```
